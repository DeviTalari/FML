```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,10,100)
x
```

```python
plt.plot(x,x,label='linear')
plt.legend()
plt.show()
```

```python
plt.plot([1,2,3,4],[10,20,25,30],color='lightblue',linewidth=3)
plt.scatter([0.3,3.8,1.2,2.5],[11,25,9,26],color='darkgreen',marker='^')
plt.xlim(0.5,4.5)
plt.show()
```

```python
x = np.array(["a","b","c","d"])
y = np.array([3,8,1,10])
plt.bar(x,y)

plt.show()
```

```python
x = np.array(["a","b","c","d"])
y = np.array([3,8,1,10])
plt.barh(x,y)

plt.show()
```
